```python
# Nettoyage des valeurs aberrantes avec IQR

def remove_outliers_iqr(df, column):
    """
    Supprime les outliers d'une colonne donnée d'un DataFrame
    selon la méthode de l'IQR (InterQuartile Range).
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    cleaned_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    print(f"Nettoyage '{column}' :")
    print(f"- Q1 : {Q1:.2f}")
    print(f"- Q3 : {Q3:.2f}")
    print(f"- IQR : {IQR:.2f}")
    print(f"- Limite basse : {lower_bound:.2f}")
    print(f"- Limite haute : {upper_bound:.2f}")
    print(f"- Lignes avant : {len(df)}")
    print(f"- Lignes après : {len(cleaned_df)}\n")

    return cleaned_df

# Nettoyage successif des 3 colonnes
df_paris = remove_outliers_iqr(df_paris, 'prix_m2')
df_paris = remove_outliers_iqr(df_paris, 'valeur_fonciere')
df_paris = remove_outliers_iqr(df_paris, 'surface_reelle_bati')
```

```python
target_name = "prix_m2"
X = df_paris.drop(target_name, axis=1)
Y = df_paris[target_name]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

rfc = RandomForestRegressor(max_depth=10)
rfc.fit(X_train, Y_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)

# Modèle 1 : Random Forest régularisé
rfc = RandomForestRegressor(
    max_depth=10,
    min_samples_leaf=5,
    random_state=42
)
rfc.fit(X_train, Y_train)
y_pred_rf = rfc.predict(X_test)

# Modèle 2 : Régression linéaire
lr = LinearRegression()
lr.fit(X_train, Y_train)
y_pred_lr = lr.predict(X_test)

# Évaluation
print("📌 Random Forest")
print("R²:", r2_score(Y_test, y_pred_rf))
print("MAE:", mean_absolute_error(Y_test, y_pred_rf))

print("\n📌 Régression Linéaire")
print("R²:", r2_score(Y_test, y_pred_lr))
print("MAE:", mean_absolute_error(Y_test, y_pred_lr))
```
